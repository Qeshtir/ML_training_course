```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# коорджинаты по X - индексы чисел
# создаём полотно
fig = plt.figure()
fig.set_size_inches(10, 8)

plt.plot([1,4,9,16]) # создаем график, где по оси X будут идти индексы элементов списка, а по оси Y - элементы списка
plt.scatter([5,7], [7,8]) # создание графика точек с координатами X и Y

plt.xlabel('Time', fontsize = 12, color = 'black') # задать параметры оси X
plt.ylabel('Millions', fontsize = 8, color = 'black') # задать параметры оси Y
plt.legend(['Salary', 'Salary in Data Science', 'Just dots']) # добавить легенду
plt.tick_params(colors='green', labelsize=8, length=10, width=2)
plt.show() # показать график
```

```python
fig = plt.figure()
fig.set_size_inches(4, 2)

x = np.linspace(-10, 10, 150) # заполняет пробелы между числами отрезками равной длины для 150 чисел от -10 до 10
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
plt.subplot(2, 2, 1) # делим полотно на части, определяем начало первой части
plt.plot(x, y)
plt.subplot(2, 2, 2) # определяем начало второй части
plt.plot(x, y)
plt.subplot(2, 1, 2) # меняем количество разрезов канваса
plt.plot(x, y)
fig.tight_layout() # задать расстояние между графиками


plt.show()
```
