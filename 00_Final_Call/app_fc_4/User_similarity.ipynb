```python
import numpy as np
import pandas as pd
import heapq
```

```python
data_user_fin = pd.read_csv('C:\\Users\\realn\\PycharmProjects\\ML_training_course\\00_Final_Call\\app_fc_2\\data_user_fin.csv')
data_user_fin
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data_user_fin_cos = data_user_fin.drop(columns=["user_id"])
```

```python
c_sim = cosine_similarity(data_user_fin_cos.iloc[[0]], data_user_fin_cos.drop([0], axis=0)).tolist()[0]
```

```python
N = 3
res = [c_sim.index(i) for i in heapq.nlargest(N, c_sim)]
res = res[::-1]
 
# printing result
print("Indices list of max N elements is : " + str(res))
```

```python
data_user_fin.iloc[res]["user_id"].tolist()
```

```python
data_user_fin_ = data_user_fin.copy()
```

```python
data_user_fin_["users_1"] = 0
data_user_fin_["users_2"] = 0
data_user_fin_["users_3"] = 0
```

```python
data_user_fin_
```

```python
data_user_fin_.loc[0, "users_1"] = data_user_fin.iloc[res]["user_id"].tolist()[0]
data_user_fin_.loc[0, "users_2"] = data_user_fin.iloc[res]["user_id"].tolist()[1]
data_user_fin_.loc[0, "users_3"] = data_user_fin.iloc[res]["user_id"].tolist()[2]
```

```python
data_user_fin_
```

```python
from tqdm import tqdm
def cosine_sim_users():
    for user in tqdm(range(len(data_user_fin)), desc="cosine"):
        c_sim = cosine_similarity(data_user_fin_cos.iloc[[user]], data_user_fin_cos.drop([user], axis=0)).tolist()[0]
        res = [c_sim.index(i) for i in heapq.nlargest(N, c_sim)]
        res = res[::-1]
        data_user_fin_.loc[user, "users_1"] = data_user_fin.iloc[res]["user_id"].tolist()[0]
        data_user_fin_.loc[user, "users_2"] = data_user_fin.iloc[res]["user_id"].tolist()[1]
        data_user_fin_.loc[user, "users_3"] = data_user_fin.iloc[res]["user_id"].tolist()[2]
```

```python
cosine_sim_users()

data_user_fin_
```

```python
data_post_fin = pd.read_csv('C:\\Users\\realn\\PycharmProjects\\ML_training_course\\00_Final_Call\\app_fc_4\\data_post_fin_nn.csv')
```

```python
like_table = pd.read_csv('C:\\Users\\realn\\PycharmProjects\\ML_training_course\\00_Final_Call\\app_fc_2\\like_table_ts.csv')
```

```python
from catboost import CatBoostClassifier
from_file = CatBoostClassifier()
from_file.load_model('C:\\Users\\realn\\PycharmProjects\\ML_training_course\\00_Final_Call\\app_fc_4\\cbm_best_balanced.cbm', format='cbm')
```

```python
users_table = data_user_fin_
```

```python
user_id = 200
limit = 5
```

```python
%time
# Будем уверенными, что переданная в методе дата подходит для выбранного нами формата
f = pd.to_datetime("1990-01-01")
f = (f - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s')


# Найдём юзера
test_user = users_table[users_table['user_id'] == user_id]

# Найдём посты, которые нужно для юзера исключить
excluded_posts = like_table[like_table['user_id'] == user_id]
excluded_posts = excluded_posts[excluded_posts['timestamp'] < f]['post_id']

# Исключим посты из общего пула
posts_fin = data_post_fin.merge(excluded_posts, indicator=True, how='left').loc[
    lambda x: x['_merge'] != 'both'].drop(columns=['_merge'])

# Подмержим юзера, нужно для предикта
user_fin = pd.merge(posts_fin, test_user, how="cross").drop(columns=["users_1", "users_2", "users_3"])

# Возьмём похожих юзеров
user_1 = users_table[users_table['user_id'] == int(test_user["users_1"])].drop(columns=["users_1", "users_2", "users_3"])
user_2 = users_table[users_table['user_id'] == int(test_user["users_2"])].drop(columns=["users_1", "users_2", "users_3"])
user_3 = users_table[users_table['user_id'] == int(test_user["users_3"])].drop(columns=["users_1", "users_2", "users_3"])

# Подмержим к ним посты
user_1_fin = pd.merge(posts_fin, user_1, how="cross")
user_2_fin = pd.merge(posts_fin, user_2, how="cross")
user_3_fin = pd.merge(posts_fin, user_3, how="cross")

# Сделаем предикт
user_probas = pd.concat([user_fin,
                             pd.Series(from_file.predict_proba(user_fin)[:, 1], index=user_fin.index, name='probas')],
                            axis=1)

# Также предикты для похожих
user_1_probas = pd.concat([user_fin,
                             pd.Series(from_file.predict_proba(user_1_fin)[:, 1], index=user_fin.index, name='probas')],
                            axis=1)
user_2_probas = pd.concat([user_fin,
                             pd.Series(from_file.predict_proba(user_2_fin)[:, 1], index=user_fin.index, name='probas')],
                            axis=1)
user_3_probas = pd.concat([user_fin,
                             pd.Series(from_file.predict_proba(user_3_fin)[:, 1], index=user_fin.index, name='probas')],
                            axis=1)

# Выберем топ limit для основного, вернём датафреймом
post_df = user_probas.sort_values('probas', ascending=False).iloc[:limit][['post_id', 'probas']]

# То же самое для похожих
post_df_1 = user_1_probas.sort_values('probas', ascending=False).iloc[:limit][['post_id', 'probas']]
post_df_2 = user_2_probas.sort_values('probas', ascending=False).iloc[:limit][['post_id', 'probas']]
post_df_3 = user_3_probas.sort_values('probas', ascending=False).iloc[:limit][['post_id', 'probas']]


# Соединим датасеты
res_df = pd.concat([post_df, post_df_1, post_df_2, post_df_3], ignore_index=True)

# Отсортируем по вероятности
res_df = res_df.sort_values('probas', ascending=False)

# Сбросим дубли
res_df = res_df.drop_duplicates(subset=["post_id"], keep='first')

# вернём список
post_list = list(res_df[:limit]['post_id'])
```
